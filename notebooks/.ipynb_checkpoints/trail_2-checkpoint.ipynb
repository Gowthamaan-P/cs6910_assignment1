```python
# Load all necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
import wandb
from sklearn.model_selection import train_test_split
```

```python
#Random seed to initiate initial weights and bias. 
#This decides the overall convergence of the model. 
np.random.seed(27) 
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Flatten the images
train_images = train_images.reshape(train_images.shape[0], 784) / 255

# Encode the labels
train_labels = np.eye(10)[train_labels]

# Prepare data for training and validation
images_train, images_val, labels_train, labels_val = train_test_split(train_images, train_labels, test_size=0.1, shuffle=True, random_state=27)
```

```python
class Activation:
    def __init__(self):
        pass
    
    def activation(self, activation_function, x):
        if activation_function == "sigmoid":
            return 1 / (1 + np.exp(-x))
        elif activation_function == "tanh":
            return np.tanh(x)
        elif activation_function == "relu":
            return np.maximum(0, x)
        return x
    
    def output_activation(self,output_activation_function,  x):
        if output_activation_function == "softmax":
            max_x = np.max(x, axis=1)
            max_x = max_x.reshape(max_x.shape[0], 1)
            exp_x = np.exp(x - max_x)
            softmax_mat = exp_x / np.sum(exp_x, axis=1).reshape(exp_x.shape[0], 1)
            return softmax_mat
        
    def activation_derivative(self, activation_function, x):
        if activation_function == "sigmoid":
            return x * (1 - x)
        elif activation_function == "tanh":
            return 1 - x ** 2
        elif activation_function == "relu":
            return np.where(x > 0, 1, 0)
        
    def output_activation_derivative(self, output_activation_function, y_pred):
        if output_activation_function == "softmax":
            return np.diag(y_pred) - np.outer(y_pred, y_pred)
```

```python
class Criterion:
    def __init__(self):
        pass
    
    def loss(self, criterion, y, y_pred):
        if criterion == "cel":
            return -np.sum(y * np.log(y_pred))
        elif criterion == "mse":
            return np.sum((y - y_pred) ** 2) / 2
        
    def loss_derivative(self, criterion, y, y_pred):
        if criterion == "cel":
            return -y / y_pred
        elif criterion == "mse":
            return (y_pred - y)
```

```python
class NeuralNetwork:
    def __init__(self, 
                neurons=128, 
                hidden_layers=4, 
                input_size=784, 
                output_size=10, 
                activation="sigmoid", 
                weight_initialization="random",
                criterion="cel", 
                output_activation="softmax"):
                
        self.neurons, self.hidden_layers = neurons, hidden_layers
        self.weights, self.biases = [], []
        self.input_size, self.output_size = input_size, output_size
        self.activation_function, self.weight_init_method = activation, weight_initialization
        self.output_activation_function = output_activation
        self.criterion = criterion

        # Initialize weights
        self.initialize_weights()
        self.initiate_biases()
        
        self.g = Activation()
        self.c = Criterion()
    
    def initialize_weights(self):
        self.weights.append(np.random.randn(self.input_size, self.neurons))
        for _ in range(self.hidden_layers - 1):
            self.weights.append(np.random.randn(self.neurons, self.neurons))
        self.weights.append(np.random.randn(self.neurons, self.output_size))

        if self.weight_init_method == "xavier":
            for i in range(len(self.weights)):
                self.weights[i] = self.weights[i] * np.sqrt(1 / self.weights[i].shape[0])

    def initiate_biases(self):
        for _ in range(self.hidden_layers):
            self.biases.append(np.zeros(self.neurons))
        self.biases.append(np.zeros(self.output_size))
    
    def forward(self, x):
        self.linear_combination, self.activation = [x], [x]
        
        # Hidden layers
        for i in range(self.hidden_layers):            
            self.linear_combination.append(np.matmul(self.activation[-1], self.weights[i]) + self.biases[i])
            self.activation.append(
                self.g.activation(self.activation_function, self.linear_combination[-1])
            )
        
        # Output layer (last layer)
        self.linear_combination.append(np.matmul(self.activation[-1], self.weights[-1]) + self.biases[-1])
        self.activation.append(
            self.g.output_activation(self.output_activation_function, self.linear_combination[-1])
        )

        return self.activation[-1]
    
    def backward(self, y, y_pred):
        self.weights_update, self.biases_update = [], []
        self.error, self.delta = [], []

        self.error.append(
            self.c.loss_derivative(self.criterion, y, y_pred)
        )
        
        output_derivative_matrix = []
        for i in range(y_pred.shape[0]):
            output_derivative_matrix.append(np.matmul(
                self.c.loss_derivative(self.criterion, y[i], y_pred[i]), 
                self.g.output_activation_derivative(self.output_activation_function, y_pred[i])
            ))
        self.delta.append(np.array(output_derivative_matrix))

        for i in range(self.hidden_layers, 0, -1):
            self.weights_update.append(np.matmul(self.activation[i].T, self.delta[-1]))
            self.biases_update.append(np.sum(self.delta[-1], axis=0))
            self.error.append(np.matmul(self.delta[-1], self.weights[i].T))
            self.delta.append(self.error[-1] * self.g.activation_derivative(self.activation_function, self.activation[i]))

        self.weights_update.append(np.matmul(self.activation[0].T, self.delta[-1]))
        self.biases_update.append(np.sum(self.delta[-1], axis=0))

        self.weights_update.reverse()
        self.biases_update.reverse()
        for i in range(len(self.weights_update)):
            self.weights_update[i] = self.weights_update[i] / y.shape[0]
            self.biases_update[i] = self.biases_update[i] / y.shape[0]

        return self.weights_update, self.biases_update
```

```python
class Optimizer:
    def __init__(self, 
                 neural_net: NeuralNetwork, 
                 lr=0.001, 
                 optimizer="sgd", 
                 beta=0.9,
                 epsilon=1e-8,
                 beta1=0.9,
                 beta2=0.999, 
                 timestep=0,
                 decay=0):
        
        # Initialize parameters
        self.neural_net, self.lr, self.optimizer = neural_net, lr, optimizer
        self.beta, self.epsilon, self.beta1, self.beta2= beta, epsilon, beta1, beta2
        self.hidden_weights = [np.zeros_like(w) for w in self.neural_net.weights]
        self.hidden_biases = [np.zeros_like(b) for b in self.neural_net.biases]
        self.m_hidden_weights = [np.zeros_like(w) for w in self.neural_net.weights]
        self.m_hidden_biases = [np.zeros_like(b) for b in self.neural_net.biases]
        self.timestep = timestep
        self.decay = decay

    def optimize(self, weights, biases):
        if(self.optimizer == "sgd"):
            self.sgd(weights, biases)
        elif(self.optimizer == "momentum"):
            self.momentum(weights, biases)
        elif(self.optimizer == "nag"):
            self.nesterov(weights, biases)
        elif(self.optimizer == "rmsprop"):
            self.rms_prop(weights, biases)
        elif(self.optimizer == "adam"):
            self.adam(weights, biases)
        elif (self.optimizer == "nadam"):
            self.nadam(weights, biases)
    
    def sgd(self, weights, biases):
        for i in range(self.neural_net.hidden_layers + 1):
            self.neural_net.weights[i] -= self.lr * (weights[i] + self.decay * self.neural_net.weights[i])
            self.neural_net.biases[i] -= self.lr * (biases[i] + self.decay * self.neural_net.biases[i])

    def momentum(self, weights, biases):
        for i in range(self.neural_net.hidden_layers + 1):
            self.hidden_weights[i] = self.beta * self.hidden_weights[i] + weights[i]
            self.hidden_biases[i] = self.beta * self.hidden_biases[i] + biases[i]

            self.neural_net.weights[i] -= self.lr * (self.hidden_weights[i] + self.decay * self.neural_net.weights[i])
            self.neural_net.biases[i] -= self.lr * (self.hidden_biases[i] + self.decay * self.neural_net.biases[i])

    def nesterov(self, weights, biases):        
        for i in range(self.neural_net.hidden_layers + 1):
            self.hidden_weights[i] = self.beta * self.hidden_weights[i] + weights[i]
            self.hidden_biases[i] = self.beta * self.hidden_biases[i] + biases[i]

            self.neural_net.weights[i] -= self.lr * (self.beta * self.hidden_weights[i] + weights[i] + self.decay * self.neural_net.weights[i])
            self.neural_net.biases[i] -= self.lr * (self.beta * self.hidden_biases[i] + biases[i] + self.decay * self.neural_net.biases[i])

    def rms_prop(self, weights, biases):
        for i in range(self.neural_net.hidden_layers + 1):
            self.hidden_weights[i] = self.beta * self.hidden_weights[i] + (1 - self.beta) * weights[i]**2
            self.hidden_biases[i] = self.beta * self.hidden_biases[i] + (1 - self.beta) * biases[i]**2

            self.neural_net.weights[i] -= (self.lr / (np.sqrt(self.hidden_weights[i]) + self.epsilon)) * weights[i] + self.decay * self.neural_net.weights[i] * self.lr
            self.neural_net.biases[i] -= (self.lr / (np.sqrt(self.hidden_biases[i]) + self.epsilon)) * biases[i] + self.decay * self.neural_net.biases[i] * self.lr

    def adam(self, d_weights, d_biases):
        for i in range(self.neural_net.hidden_layers + 1):
            self.m_hidden_weights[i] = self.beta1 * self.m_hidden_weights[i] + (1 - self.beta1) * d_weights[i]
            self.m_hidden_biases[i] = self.beta1 * self.m_hidden_biases[i] + (1 - self.beta1) * d_biases[i]

            self.hidden_weights[i] = self.beta2 * self.hidden_weights[i] + (1 - self.beta2) * d_weights[i]**2
            self.hidden_biases[i] = self.beta2 * self.hidden_biases[i] + (1 - self.beta2) * d_biases[i]**2

            self.corrected_hidden_weights_b1 = self.m_hidden_weights[i] / (1 - self.beta1**(self.timestep + 1))
            self.corrected_hidden_biases_b1 = self.m_hidden_biases[i] / (1 - self.beta1**(self.timestep + 1))

            self.corrected_hidden_weights_b2 = self.hidden_weights[i] / (1 - self.beta2**(self.timestep + 1))
            self.corrected_hidden_biases_b2 = self.hidden_biases[i] / (1 - self.beta2**(self.timestep + 1))

            self.neural_net.weights[i] -= self.lr * (self.corrected_hidden_weights_b1 / ((np.sqrt(self.corrected_hidden_weights_b2)) + self.epsilon)) + self.decay * self.neural_net.weights[i] * self.lr
            self.neural_net.biases[i] -= self.lr * (self.corrected_hidden_biases_b1 / ((np.sqrt(self.corrected_hidden_biases_b2)) + self.epsilon)) + self.decay * self.neural_net.biases[i] * self.lr

    def nadam(self, d_weights, d_biases):
        for i in range(self.neural_net.hidden_layers + 1):
            self.m_hidden_weights[i] = self.beta1 * self.m_hidden_weights[i] + (1 - self.beta1) * d_weights[i]
            self.m_hidden_biases[i] = self.beta1 * self.m_hidden_biases[i] + (1 - self.beta1) * d_biases[i]

            self.hidden_weights[i] = self.beta2 * self.hidden_weights[i] + (1 - self.beta2) * d_weights[i]**2
            self.hidden_biases[i] = self.beta2 * self.hidden_biases[i] + (1 - self.beta2) * d_biases[i]**2

            self.corrected_hidden_weights_b1 = self.m_hidden_weights[i] / (1 - self.beta1 ** (self.timestep + 1))
            self.corrected_hidden_biases_b1 = self.m_hidden_biases[i] / (1 - self.beta1 ** (self.timestep + 1))

            self.corrected_hidden_weights_b2 = self.hidden_weights[i] / (1 - self.beta2 ** (self.timestep + 1))
            self.corrected_hidden_biases_b2 = self.hidden_biases[i] / (1 - self.beta2 ** (self.timestep + 1))

            temp_update_w = self.beta1 * self.corrected_hidden_weights_b1 + ((1 - self.beta1) / (1 - self.beta1 ** (self.timestep + 1))) * d_weights[i]
            temp_update_b = self.beta1 * self.corrected_hidden_biases_b1 + ((1 - self.beta1) / (1 - self.beta1 ** (self.timestep + 1))) * d_biases[i]

            self.neural_net.weights[i] -= self.lr * (temp_update_w / ((np.sqrt(self.corrected_hidden_weights_b2)) + self.epsilon)) + self.decay * self.neural_net.weights[i] * self.lr
            self.neural_net.biases[i] -= self.lr * (temp_update_b / ((np.sqrt(self.corrected_hidden_biases_b2)) + self.epsilon)) + self.decay * self.neural_net.biases[i] * self.lr
```

```python
def train(config, X_train, Y_train, X_val, Y_val, show_log): 
    train_loss_hist = []
    train_accuracy_hist = []
    val_loss_hist = []
    val_accuracy_hist = []
    
    nn = NeuralNetwork(input_size=config['input_size'], 
                       output_size=config['output_size'], 
                       hidden_layers=config['hidden_layers'], 
                       neurons=config['neurons'],  
                       activation=config['activation'], 
                       output_activation=config['output_activation'],
                       criterion=config['criterion'],
                       weight_initialization=config['weight_initialization'])
    
    optimizer = Optimizer(neural_net=nn,
                    lr=config['learning_rate'],
                    optimizer=config['optimizer'],
                    beta=config['beta'],
                    epsilon=config['epsilon'],
                    beta1=config['beta1'],
                    beta2=config['beta2'],
                    decay=config['decay'])
    
    batch_size = config['batch_size']
    c = Criterion()
    
    for epoch in range(config['epochs']):
        for batch in range(0, X_train.shape[0], batch_size):
            # Get the batch of data
            X_batch = X_train[batch:batch+batch_size]
            Y_batch = Y_train[batch:batch+batch_size]

            Y_hat_batch = nn.forward(X_batch)
            weights, biases = nn.backward(Y_batch, Y_hat_batch)
            optimizer.optimize(weights, biases)
        
        optimizer.timestep += 1
        
        # Training
        Y_hat_train = nn.forward(X_train)
        train_loss = c.loss(config['criterion'], Y_train, Y_hat_train)
        train_accuracy = np.sum(np.argmax(Y_hat_train, axis=1) == np.argmax(Y_train, axis=1)) / Y_train.shape[0]
            
        # Validation
        Y_hat_val = nn.forward(X_val)
        val_loss = c.loss(config['criterion'], Y_val, Y_hat_val)
        val_accuracy = np.sum(np.argmax(Y_hat_val, axis=1) == np.argmax(Y_val, axis=1)) / Y_val.shape[0]
        
        if show_log:
            print("Epoch {} Train Loss {} Train Accuracy {} Val Loss {} Val Accuracy {}".format(epoch, train_loss, train_accuracy, val_loss, val_accuracy))
   
        train_loss_hist.append(train_loss)
        train_accuracy_hist.append(train_accuracy)
        val_loss_hist.append(val_loss)
        val_accuracy_hist.append(val_accuracy)
    
    return nn, train_loss_hist, train_accuracy_hist, val_loss_hist, val_accuracy_hist
```

```python
# Network Config
network_config = {
    'input_size': 784,
    'output_size': 10,
    'hidden_layers': 1,
    'neurons':256,
    'activation':'sigmoid',
    'output_activation':'softmax',
    'learning_rate': 0.005,
    'beta': 0.8,
    'beta1': 0.9,
    'beta2':0.9999,
    'epsilon': 1e-8,
    'epochs': 10,
    'optimizer': "nadam",
    'criterion': "cel",
    'decay': 0.0005,
    'weight_initialization': "random",
    'batch_size': 64,
}
        
nn, tl, ta, vl, va = train(network_config, images_train, labels_train, images_val, labels_val, show_log = True)
```

```python
# Plot changes in loss
plt.plot(tl, c = 'b', label = 'Train')
plt.plot(vl, c = 'r', label = 'Validation')
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(linestyle='-.', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# Plot changes in accuracy
plt.plot(ta, c = 'b', label = 'Train')
plt.plot(va, c = 'r', label = 'Validation')
plt.title('Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuarcy')
plt.grid(linestyle='-.', linewidth=0.5)
plt.legend()
plt.show()
```
